## 11.3 Least Squares and the Fitted Model

### Exercise 11.3 (b)

```python
xi = [0] * 3 + [15] * 3 + [30] * 3 + [45] * 3 + [60] * 3 + [75] * 3
yi = [8, 6, 8, 12, 10, 14, 25, 21, 24, 31, 33, 28, 44, 39, 42, 48, 51, 44]
```

```python
print(len(xi), len(yi))
```

```python
import numpy as np

b = [5.8254, 0.56762]
x = np.linspace(0, 75, 100)
x = x.reshape((-1,1))
x = np.concatenate((np.ones(len(x)).reshape((-1,1)), x), axis=1)
y = np.dot(b, x.T)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(xi, yi)
plt.plot(x[:,1], y, color='red')
plt.show()
```

```python
# linear regression using sklearn
from sklearn import linear_model

reg = linear_model.LinearRegression()
reg.fit(np.array(xi).reshape((-1, 1)), yi)
y_pred = reg.predict(np.array(xi).reshape((-1, 1)))

plt.scatter(xi, yi)
plt.plot(xi, y_pred, color='red')
plt.show()
```

### Exercise 11.9

```python
#(a) Plot a scatter diagram.

advertising_cost = [40, 20, 25, 20, 30, 50, 40, 20, 50, 40, 25, 50]
sales = [385, 400, 395, 365, 475, 440, 490, 420, 560, 525, 480, 510]

plt.scatter(advertising_cost, sales)
plt.show()
```

```python
# (b) Find the equation of the regression line to predict weekly sales from advertising expenditures.

b = [343.7056, 3.220812]
x = np.linspace(20, 50, 100)
x = np.concatenate((np.ones(len(x)).reshape((-1,1)), x.reshape((-1,1))), axis=1)
y = np.dot(b, x.T)

plt.scatter(advertising_cost, sales)
plt.plot(x[:,1], y, color='red')
plt.xlabel('Advertising Costs')
plt.ylabel('Sales')
plt.show()
```

```python
#(d) Plot the residuals versus advertising costs.

ad_cost = np.concatenate((np.ones(len(advertising_cost)).reshape((-1,1)), np.array(advertising_cost).reshape((-1,1))), axis=1)
sales_pred = np.dot(b, ad_cost.T)
residuals = sales - sales_pred

plt.scatter(advertising_cost, residuals)
plt.show()
```
